```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pickle
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import (RBF, Matern, RationalQuadratic,
                                              ExpSineSquared, DotProduct,
                                              ConstantKernel)
import scipy
import scipy.ndimage

with open('big_o_ground_estimates.pkl', 'rb') as f:
    results = pickle.load(f)
```

```python
plt.figure(figsize=(24, 7))
plt.subplot(131)
plt.imshow(results['original'])
plt.title('Original map')
plt.subplot(132)
plt.imshow(results['coarse'])
plt.title('In-filled (coarse)')
plt.subplot(133)
plt.imshow(results['fine'])
plt.title('In-filled (fine)')
```

```python
xrg = slice(0, 128)
yrg = slice(128, 256)
plt.figure(figsize=(24, 7))
plt.subplot(131)
plt.imshow(results['original'][yrg, xrg])
plt.title('Original map')
plt.subplot(132)
plt.imshow(results['coarse'][yrg, xrg])
plt.title('In-filled (coarse)')
plt.subplot(133)
plt.imshow(results['fine'][yrg, xrg])
plt.title('In-filled (fine)')
```
